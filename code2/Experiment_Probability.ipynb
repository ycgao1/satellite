```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

def calculate_success_probability(N, k):
    """
    Calculate the probability of at least one successful attack targeting a single active item
    among N total items over k non-independent attacks.
    
    Args:
    N (int): Total number of items.
    k (int): Number of attacks.
    
    Returns:
    float: Probability of at least one success.
    """
    # Probability of failing all attacks, assuming each attack targets a different item
    if k > N:  # More attacks than items
        return 1.0  # Guaranteed to hit the item
    p_fail = 1.0
    for i in range(k):
        p_fail *= (N - 1 - i) / (N - i)
    p_success = 1 - p_fail
    return p_success

def plot_success_probability_multiple_N_values(N_values, k_max):
    """
    Plot the probability of at least one successful attack for multiple values of N,
    where each line represents a different total number of items.
    
    Args:
    N_values (list of int): List of total item numbers.
    k_max (int): Maximum number of attacks.
    """
    k_range = range(1, k_max + 1)
    plt.figure(figsize=(12, 8))

    for N in N_values:
        probabilities = [calculate_success_probability(N, k) for k in k_range]
        plt.plot(k_range, probabilities, 'o-', label=f'N={N}')

    plt.title("Probability of At Least One Successful Attack for Various N")
    plt.xlabel("Number of Attacks (k)")
    plt.ylabel("Probability of Success")
    plt.legend()
    plt.grid(True)
    plt.gca().yaxis.set_major_formatter(PercentFormatter(1))  # Formatting the y-axis as percentage
    plt.show()

# List of N values to plot
N_values = [50, 100, 150, 200, 250]
k_max = 10  # Maximum number of attacks

# Generate the plot for multiple N values
plot_success_probability_multiple_N_values(N_values, k_max)
```

```python
from math import comb
import matplotlib.pyplot as plt

def probability_all_important_detected(N, n, k):
    """
    Calculate the probability that all n important items are detected within k attacks out of N total items.
    
    Args:
    N (int): Total number of items.
    n (int): Number of important items.
    k (int): Number of attacks.
    
    Returns:
    float: Probability that all important items are detected.
    """
    if k < n:
        return 0  # Not enough attacks to possibly detect all important items
    return comb(N - n, k - n) / comb(N, k)

def plot_probability_all_important_detected(N, n_values, k_max):
    """
    Plot the probability of detecting all important items for different values of n,
    given a fixed N, over a range of k values.
    
    Args:
    N (int): Total number of items.
    n_values (list of int): List of number of important items.
    k_max (int): Maximum number of attacks.
    """
    k_range = range(1, k_max + 1)
    plt.figure(figsize=(10, 6))

    n = 1
    probabilities = [probability_all_important_detected(N, n, k) if k >= n else 0 for k in k_range]
    plt.plot(k_range, probabilities, 'o-', label=f'n={n} Stealthy Deployment')

    n = 2
    probabilities = [probability_all_important_detected(N, n, k) if k >= n else 0 for k in k_range]
    plt.plot(k_range, probabilities, 'o-', label=f'n={n} Dynamic Re-Routing' )

    n = 3
    probabilities = [probability_all_important_detected(N, n, k) if k >= n else 0 for k in k_range]
    plt.plot(k_range, probabilities, 'o-', label=f'n={n} Dynamic Re-Routing')

    n = 4
    probabilities = [probability_all_important_detected(N, n, k) if k >= n else 0 for k in k_range]
    plt.plot(k_range, probabilities, 'o-', label=f'n={n} Dynamic Re-Routing')
    

    #plt.title(f"Probability of Detecting All Important Items for N={N}")
    plt.xlabel("Number of Attacks")
    plt.ylabel("Attack Success Rate")
    plt.legend()
    plt.grid(True)
    plt.gca().yaxis.set_major_formatter(PercentFormatter(1))  # Formatting the y-axis as percentage
    plt.show()

# Parameters for the specific plot
N = 50  # Total number of items
n_values = [1, 2, 3]  # Number of important items
k_max = 10  # Maximum number of attacks

# Generate the plot with specific parameters
plot_probability_all_important_detected(N, n_values, k_max)
```

```python
def plot_probability_vs_N_for_fixed_k(N_values, n_values, k_fixed):
    """
    Plot the probability of detecting all important items for a fixed number of attacks (k),
    over a range of N values, and for different values of n.
    
    Args:
    N_values (list of int): List of total item numbers.
    n_values (list of int): List of number of important items.
    k_fixed (int): Fixed number of attacks.
    """
    plt.figure(figsize=(10, 6))

    n = 1
    probabilities = [probability_all_important_detected(N, n, k_fixed) if k_fixed >= n else 0 for N in N_values]
    plt.plot(N_values, probabilities, 'o-', label=f'n={n} Stealthy Deployment')

    n = 2
    probabilities = [probability_all_important_detected(N, n, k_fixed) if k_fixed >= n else 0 for N in N_values]
    plt.plot(N_values, probabilities, 'o-', label=f'n={n} Dynamic Re-Routing')

    n = 3
    probabilities = [probability_all_important_detected(N, n, k_fixed) if k_fixed >= n else 0 for N in N_values]
    plt.plot(N_values, probabilities, 'o-', label=f'n={n} Dynamic Re-Routing')

    n = 4
    probabilities = [probability_all_important_detected(N, n, k_fixed) if k_fixed >= n else 0 for N in N_values]
    plt.plot(N_values, probabilities, 'o-', label=f'n={n} Dynamic Re-Routing')

    #plt.title(f"Probability of Detecting All Important Items for k={k_fixed}")
    plt.xlabel("Total Number of Satellites")
    plt.ylabel("Attack Success Rate")
    plt.xticks(N_values)  # Set x-ticks to be exactly the values of N
    plt.legend()
    plt.grid(True)
    plt.gca().yaxis.set_major_formatter(PercentFormatter(1))  # Formatting the y-axis as percentage
    plt.show()

# Parameters for the specific plot
N_values = [25, 50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300]  # Different total number of items
n_values = [1, 2, 3]  # Number of important items
k_fixed = 5  # Fixed number of attacks

# Generate the plot with specific parameters
plot_probability_vs_N_for_fixed_k(N_values, n_values, k_fixed)
```
